THEORY QUESTIONS

1) What is a random variable in probability theory?

- In probability theory, a random variable is a numerical outcome of a random process or experiment. It assigns a number to each outcome in the sample space of a probability experiment.

2) What are the types of random variables?

Discrete Random Variable
- Takes countable values (finite or countably infinite).
- Example: Number of heads in 3 coin tosses → values: 0, 1, 2, 3.

Continuous Random Variable
- Takes uncountable values (typically any value in an interval).
- Example: Time it takes to run a marathon → values: any real number ≥ 0.

3) What is the difference between discrete and continuous distributions?

Discrete Distribution
- Random Variable Type is Discrete.
- Values	Countable (0, 1, 2, …)
- Function Used	is PMF (Probability Mass)
- Example-	Binomial, Poisson

Continuous Distribution
- Random Variable Type is contiuous.
- Uncountable (real numbers)
- Function used is PDF (Probability Density)
- Example- Normal, Exponential

4) What are probability distribution functions (PDF)?

- In probability theory, a Probability Distribution Function (PDF) describes how probabilities are distributed over the values of a random variable.
- The meaning of PDF depends on whether the variable is discrete or continuous.

5) How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)?

PDF / PMF
- Probability at a point (PMF), or density (PDF)
- Shapes are Bar graph (PMF), smooth curve (PDF)
- Range	PDF: ≥ 0; area = 1
PMF: values ≤ 1
- Interpretation is Instantaneous likelihood

CDF
- Cumulative probability up to a point.
- Shapes are Smooth increasing curve.
- Range- Always between 0 and 1
- Interpretation is Total probability up to x.

6) What is a discrete uniform distribution?

- A discrete uniform distribution is a probability distribution where each outcome in a finite set of values has an equal probability of occurring.

7) What are the key properties of a Bernoulli distribution?

- PMF (Probability Mass Function)
- Mean (Expected Value)
- Variance
- Skewness
- Kurtosis

8) What is the binomial distribution, and how is it used in probability?

- The binomial distribution models the number of successes in a fixed number of independent and identically distributed (i.i.d.) Bernoulli trials, where each trial has only two outcomes: success (with probability p) or failure (with probability 1-p).

9) What is the Poisson distribution and where is it applied?

The Poisson distribution is a discrete probability distribution that describes the number of times an event occurs in a fixed interval of time or space, given that:

- Events occur independently
- The average rate of occurrence (λ) is constant.
- Two events cannot occur at the exact same instant

10) What is a continuous uniform distribution?

- The continuous uniform distribution is a probability distribution where all values within a given interval are equally likely. It is also called the rectangular distribution because of its constant probability density over the interval.

11) What are the characteristics of a normal distribution?

The normal distribution, also called the Gaussian distribution, is a continuous, symmetric, bell-shaped distribution that is widely used in statistics and data science. It models many natural phenomena like heights, test scores, measurement errors, etc.

12) What is the standard normal distribution, and why is it important?

- The standard normal distribution is a special case of the normal distribution where:

- Mean (μ=0)
- Standard deviation (σ=1)
- It is denoted by the random variable Z
- This distribution is also called the Z-distribution.

13) What is the Central Limit Theorem (CLT), and why is it critical in statistics?

- The Central Limit Theorem (CLT) is one of the most important and powerful concepts in statistics. It states:

If you take a large number of independent random samples from any population (with finite mean and variance), the distribution of the sample means will approximate a normal distribution — regardless of the shape of the original population distribution.

14) How does the Central Limit Theorem relate to the normal distribution?

- The Central Limit Theorem (CLT) explains why and when the normal distribution appears so frequently in statistics, even if the underlying data is not normally distributed.
- The CLT states that the sampling distribution of the sample mean becomes approximately normal as the sample size increases, regardless of the shape of the original population distribution.

15) What is the application of Z statistics in hypothesis testing?

Z-statistics (also called Z-scores) are used in hypothesis testing when:
- The population standard deviation σ is known.
- The sample size is large (typically n≥30)
- It helps test whether a sample mean, proportion, or difference between means is significantly different from a hypothesized population value.

16) How do you calculate a Z-score, and what does it represent?

- A Z-score (or standard score) tells you how many standard deviations a data point is from the mean of a distribution.
- It helps standardize values from different scales and compare them directly.

17) What are point estimates and interval estimates in statistics?

- Both point estimates and interval estimates are used to infer population parameters (like the mean, proportion, variance) from a sample.

18) What is the significance of confidence intervals in statistical analysis?

- A confidence interval (CI) provides a range of values within which we expect the true population parameter (like mean or proportion) to lie, with a certain level of confidence (e.g., 95%).

19) What is the relationship between a Z-score and a confidence interval?

- The Z-score is directly used in constructing confidence intervals when the population standard deviation is known and the sample size is sufficiently large.
- A confidence interval is calculated using a Z-score (also called the critical value) from the standard normal distribution.

20) How are Z-scores used to compare different distributions?

- Z-scores allow you to standardize values from different distributions, making them directly comparable — even if the original data had different means and standard deviations.

21) What are the assumptions for applying the Central Limit Theorem?

-	Random Sampling-	The data should be collected using a random process. Every member of the population must have an equal chance of being included.
- Independence-	Each observation in the sample must be independent of the others. One outcome should not influence another.
- Identically Distributed- All observations should come from the same population distribution (i.e., same mean and variance).
-	Finite Mean and Variance	The population should have a finite mean μ and finite variance.
- Sample Size- A large sample size is needed, especially if the population is not normal

22) What is the concept of expected value in a probability distribution?

- The expected value (also called mean or mathematical expectation) is the long-run average or central value of a random variable in a probability distribution.
- It represents the value you would expect to occur on average if an experiment were repeated infinitely many times.

23) How does a probability distribution relate to the expected outcome of a random variable?

- A probability distribution describes how likely each outcome of a random variable is. The expected outcome (or expected value) is the weighted average of all these outcomes — where each outcome is weighted by its probability.

```python
#2) Generate a discrete uniform distribution using Python and plot the probability mass function (PMF).

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Generate samples
n = 10000  # number of samples
low, high = 1, 6  # Discrete uniform distribution from 1 to 6 (inclusive)

samples = np.random.randint(low, high + 1, size=n)

# Step 2: Calculate probabilities (PMF)
values, counts = np.unique(samples, return_counts=True)
pmf = counts / n

# Step 3: Plot PMF
plt.figure(figsize=(8, 5))
sns.barplot(x=values, y=pmf, palette="viridis")
plt.title("PMF of Discrete Uniform Distribution (1 to 6)")
plt.xlabel("Outcome")
plt.ylabel("Probability")
plt.ylim(0, max(pmf) + 0.05)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
#3) Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution.

def bernoulli_pdf(x, p):
    if x not in [0, 1]:
        return 0
    if not (0 <= p <= 1):
        raise ValueError("Probability p must be between 0 and 1")
    return p if x == 1 else 1 - p

# Example usage:
print("P(X=1):", bernoulli_pdf(1, 0.6))  # Output: 0.6
print("P(X=0):", bernoulli_pdf(0, 0.6))  # Output: 0.4
```

```python
#6) Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete uniform distribution.

import numpy as np
import matplotlib.pyplot as plt

# Step 1: Define the range of the discrete uniform distribution
low = 1
high = 6
x = np.arange(low, high + 1)  # Possible values: [1, 2, 3, 4, 5, 6]

# Step 2: Calculate PMF and CDF
pmf = np.full_like(x, fill_value=1 / len(x), dtype=float)  # Equal probability for all values
cdf = np.cumsum(pmf)  # Cumulative sum to get the CDF

# Step 3: Plot CDF
plt.figure(figsize=(8, 5))
plt.step(x, cdf, where='post', color='blue', label='CDF')
plt.scatter(x, cdf, color='red')  # Plot dots at CDF values

plt.title("CDF of Discrete Uniform Distribution (1 to 6)")
plt.xlabel("x")
plt.ylabel("Cumulative Probability")
plt.ylim(0, 1.1)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.show()
```

```python
#9) Write a Python function to calculate Z-scores from a dataset and plot them.

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def calculate_and_plot_z_scores(data, plot_title="Z-score Distribution"):
    # Step 1: Calculate mean and standard deviation
    mean = np.mean(data)
    std = np.std(data)

    # Step 2: Calculate Z-scores
    z_scores = (data - mean) / std

    # Step 3: Plot Z-score distribution
    plt.figure(figsize=(8, 5))
    sns.histplot(z_scores, bins=30, kde=True, color='skyblue', edgecolor='black')
    plt.axvline(0, color='red', linestyle='--', label='Mean (Z=0)')
    plt.title(plot_title)
    plt.xlabel("Z-score")
    plt.ylabel("Frequency")
    plt.legend()
    plt.grid(axis='y', linestyle='--', alpha=0.6)
    plt.show()

    return z_scores
```

```python
#10) Implement the Central Limit Theorem (CLT) using Python for a non-normal distribution.

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Set parameters
population_size = 100000
sample_sizes = [5, 30, 100]   # Different sample sizes to test CLT
n_samples = 1000              # Number of samples to draw in each case
scale = 2                     # Scale parameter for exponential distribution

# Step 2: Generate a non-normal population (Exponential distribution)
population = np.random.exponential(scale=scale, size=population_size)

# Step 3: Plot original (non-normal) distribution
plt.figure(figsize=(15, 10))
plt.subplot(2, 2, 1)
sns.histplot(population, bins=50, kde=True, color='orange')
plt.title("Original Exponential Distribution (Highly Skewed)")
plt.xlabel("Value")
plt.ylabel("Frequency")

# Step 4: Apply CLT for different sample sizes
for idx, size in enumerate(sample_sizes, start=2):
    sample_means = []
    for _ in range(n_samples):
        sample = np.random.choice(population, size=size)
        sample_means.append(np.mean(sample))

    # Plot distribution of sample means
    plt.subplot(2, 2, idx)
    sns.histplot(sample_means, bins=40, kde=True, color='skyblue')
    plt.title(f"Distribution of Sample Means (n = {size})")
    plt.xlabel("Sample Mean")
    plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```

```python
#14) Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal distribution.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def calculate_z_score(x, mean, std):
    z = (x - mean) / std
    return z

def plot_z_on_standard_normal(z):
    x_vals = np.linspace(-4, 4, 1000)
    y_vals = norm.pdf(x_vals)

    plt.figure(figsize=(10, 5))
    plt.plot(x_vals, y_vals, label="Standard Normal Distribution", color='blue')
    plt.fill_between(x_vals, y_vals, where=(x_vals < z), color='skyblue', alpha=0.5, label='P(Z ≤ z)')
    plt.axvline(z, color='red', linestyle='--', label=f"Z = {z:.2f}")

    plt.title("Z-score on Standard Normal Distribution")
    plt.xlabel("Z-score")
    plt.ylabel("Probability Density")
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.show()

# Example Inputs
data_point = 72
mean = 60
std_dev = 8

# Calculate and plot Z-score
z_score = calculate_z_score(data_point, mean, std_dev)
print(f"Z-score for x = {data_point} (μ = {mean}, σ = {std_dev}): {z_score:.2f}")

# Plot Z-score on standard normal curve
plot_z_on_standard_normal(z_score)
```

```python
#15) Implement hypothesis testing using Z-statistics for a sample dataset.

import numpy as np
from scipy.stats import norm

# Step 1: Define sample and population data
sample_mean = 58.5
population_mean = 60
population_std = 5
sample_size = 50
alpha = 0.05

# Step 2: Calculate Z-statistic
standard_error = population_std / np.sqrt(sample_size)
z_stat = (sample_mean - population_mean) / standard_error

# Step 3: Determine critical value and p-value
z_critical = norm.ppf(alpha)  # One-tailed test (left-tailed)
p_value = norm.cdf(z_stat)

# Step 4: Print results
print(f"Z-statistic: {z_stat:.2f}")
print(f"Critical Z-value (α = {alpha}): {z_critical:.2f}")
print(f"P-value: {p_value:.4f}")

# Step 5: Conclusion
if z_stat < z_critical:
    print("Reject the null hypothesis (H₀). Battery life is significantly less than 60 hours.")
else:
    print("Fail to reject the null hypothesis (H₀). No significant evidence battery life is less than 60 hours.")
```

```python
#16) Create a confidence interval for a dataset using Python and interpret the result.

import numpy as np
import scipy.stats as stats

def confidence_interval(data, confidence=0.95):
    n = len(data)
    mean = np.mean(data)
    std_err = stats.sem(data)  # standard error = std / sqrt(n)
    t_critical = stats.t.ppf((1 + confidence) / 2, df=n-1)

    margin_of_error = t_critical * std_err
    lower_bound = mean - margin_of_error
    upper_bound = mean + margin_of_error

    print(f"Sample mean: {mean:.2f}")
    print(f"{int(confidence*100)}% Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")
    print(f"Interpretation: We are {int(confidence*100)}% confident the true population mean lies between these bounds.")

    return (lower_bound, upper_bound)

np.random.seed(0)
sample_data = np.random.normal(loc=50, scale=10, size=30)  # mean=50, std=10, n=30

confidence_interval(sample_data, confidence=0.95)
```

```python
#17) Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean.

import numpy as np
import scipy.stats as stats

# Step 1: Generate data from a normal distribution
np.random.seed(42)
mu = 100         # true mean
sigma = 15       # true standard deviation
n = 40           # sample size

sample = np.random.normal(loc=mu, scale=sigma, size=n)

# Step 2: Calculate sample statistics
sample_mean = np.mean(sample)
sample_std = np.std(sample, ddof=1)  # sample standard deviation
standard_error = sample_std / np.sqrt(n)

# Step 3: Calculate 95% confidence interval using t-distribution
confidence = 0.95
alpha = 1 - confidence
df = n - 1
t_critical = stats.t.ppf(1 - alpha / 2, df)

margin_of_error = t_critical * standard_error
lower_bound = sample_mean - margin_of_error
upper_bound = sample_mean + margin_of_error

# Step 4: Print results
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Sample Standard Deviation: {sample_std:.2f}")
print(f"95% Confidence Interval for the Mean: ({lower_bound:.2f}, {upper_bound:.2f})")
print("Interpretation:")
print("We are 95% confident that the true population mean lies within this interval.")
```

```python
#19) Use Python to calculate and interpret the cumulative distribution function (CDF) of a Poisson distribution.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Step 1: Define the Poisson distribution parameter
lambda_ = 4  # average number of events

# Step 2: Define the range of x values
x = np.arange(0, 15)

# Step 3: Calculate CDF values
cdf_vals = poisson.cdf(x, mu=lambda_)

# Step 4: Print CDF for a specific value
k = 6
cdf_k = poisson.cdf(k, mu=lambda_)
print(f"P(X ≤ {k}) for Poisson(λ={lambda_}): {cdf_k:.4f}")

# Step 5: Plot the CDF
plt.figure(figsize=(8, 5))
plt.step(x, cdf_vals, where='post', label='Poisson CDF', color='green')
plt.scatter(x, cdf_vals, color='black')
plt.axhline(cdf_k, color='red', linestyle='--', label=f'P(X ≤ {k}) = {cdf_k:.2f}')
plt.axvline(k, color='red', linestyle='--')
plt.title(f'Cumulative Distribution Function of Poisson(λ={lambda_})')
plt.xlabel('Number of Events (x)')
plt.ylabel('Cumulative Probability')
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.show()
```

```python
#21) Write a Python program to compare the standard deviations of two datasets and visualize the difference.

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Create two datasets with different spreads
np.random.seed(0)
data1 = np.random.normal(loc=50, scale=5, size=1000)   # Smaller std dev
data2 = np.random.normal(loc=50, scale=15, size=1000)  # Larger std dev

# Step 2: Calculate standard deviations
std1 = np.std(data1, ddof=1)
std2 = np.std(data2, ddof=1)

# Step 3: Print the standard deviations
print(f"Standard Deviation of Dataset 1: {std1:.2f}")
print(f"Standard Deviation of Dataset 2: {std2:.2f}")

# Step 4: Plot both histograms
plt.figure(figsize=(10, 6))
sns.histplot(data1, bins=30, color='skyblue', label=f'Dataset 1 (std={std1:.2f})', kde=True)
sns.histplot(data2, bins=30, color='salmon', label=f'Dataset 2 (std={std2:.2f})', kde=True)

# Final touches
plt.title("Comparison of Standard Deviations between Two Datasets")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
#22) Calculate the range and interquartile range (IQR) of a dataset generated from a normal distribution.

import numpy as np

# Step 1: Generate data from a normal distribution
np.random.seed(42)
mu = 100       # mean
sigma = 15     # standard deviation
n = 1000       # sample size

data = np.random.normal(loc=mu, scale=sigma, size=n)

# Step 2: Calculate Range
data_range = np.max(data) - np.min(data)

# Step 3: Calculate IQR (Interquartile Range)
q1 = np.percentile(data, 25)
q3 = np.percentile(data, 75)
iqr = q3 - q1

# Step 4: Print results
print(f"Range: {data_range:.2f}")
print(f"Q1 (25th percentile): {q1:.2f}")
print(f"Q3 (75th percentile): {q3:.2f}")
print(f"Interquartile Range (IQR): {iqr:.2f}")
```

```python
#23) Implement Z-score normalization on a dataset and visualize its transformation.

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Generate sample dataset (normal distribution)
np.random.seed(42)
data = np.random.normal(loc=70, scale=15, size=1000)

# Step 2: Z-score normalization
mean = np.mean(data)
std = np.std(data)
z_scores = (data - mean) / std

# Step 3: Plot original and normalized datasets
plt.figure(figsize=(12, 5))

# Original data
plt.subplot(1, 2, 1)
sns.histplot(data, bins=30, kde=True, color='orange')
plt.title("Original Data Distribution")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.grid(True, linestyle='--', alpha=0.5)

# Normalized data (Z-scores)
plt.subplot(1, 2, 2)
sns.histplot(z_scores, bins=30, kde=True, color='skyblue')
plt.title("Z-score Normalized Distribution")
plt.xlabel("Z-score")
plt.ylabel("Frequency")
plt.grid(True, linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()

# Step 4: Optional - Print summary stats
print(f"Original Mean: {mean:.2f}, Std Dev: {std:.2f}")
print(f"Z-score Mean: {np.mean(z_scores):.2f}, Std Dev: {np.std(z_scores):.2f}")
```

```python
#24) Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal distribution.

import numpy as np
from scipy.stats import skew, kurtosis

def analyze_distribution_skewness_kurtosis(mean=0, std=1, n=1000):
    # Step 1: Generate data
    data = np.random.normal(loc=mean, scale=std, size=n)

    # Step 2: Calculate skewness and kurtosis
    skewness = skew(data)
    kurt = kurtosis(data)  # By default, this returns excess kurtosis (normal = 0)

    # Step 3: Print results
    print(f"Skewness: {skewness:.4f}")
    print(f"Kurtosis (excess): {kurt:.4f}")

    return skewness, kurt

# Example usage
analyze_distribution_skewness_kurtosis(mean=100, std=15, n=1000)
```
